```python
import pandas as pd
import numpy as nm
```

```python
#   قراءة ملفات البيانات التي قمت بتحميلها من موقع دبي بلس
df_tour = pd.read_csv("C:/Users/tf/Desktop/مشروع 1/tourism_with_trend_index.csv")

```

```python
df_tour.head()
```

```python
df_tour = df_tour.rename(columns={
    'tourism_intensity_index': 'tourism_intensity',
    'tourism_intensity_index_trend': 'tourism_activity'
})
```

```python
print(df_tour.info())
print(df_tour.head(22))

df_tour['year_month'] = df_tour['year_month'].astype(str).str.strip()
df_tour['area_name_en'] = df_tour['area_name_en'].astype(str).str.strip().str.lower()
```

```python
# فحص القيم المفقودة
missing = pd.DataFrame({
    'missing_count': df_tour.isna().sum(),
    'missing_percent': df_tour.isna().sum() / len(df_tour) * 100
})

print("=== Missing Report | Tourism ===")
print(missing)
```

```python
#  فحص القيم الرقمية
num_cols = ['hotels', 'rooms', 'POIs', 'occupancy_rate']

print("=== Descriptive Statistics | Tourism ===")
print(df_tour[num_cols].describe())
```

```python
#  تنظيف منطقي آمن
df_tour = df_tour[
    (df_tour['hotels'] >= 0) &
    (df_tour['rooms'] >= 0) &
    (df_tour['POIs'] >= 0) &
    (df_tour['occupancy_rate'].between(0, 1))
]
```

```python
#  فحص قوة مفاتيح الدمج
print("عدد الأشهر:", df_tour['year_month'].nunique())
print("عدد المناطق:", df_tour['area_name_en'].nunique())

print("أكثر المفاتيح تكرارًا:")
print(
    df_tour[['year_month', 'area_name_en']]
    .value_counts()
    .head(10)
)
```

```python
#  حفظ الملف النهائي

df_tour.to_csv(
    "C:/Users/tf/Desktop/مشروع 1/Data after cleaning/tourism_cleaned.csv",
    index=False
)
```
