```python
import pandas as pd
import numpy as np
```

```python
#  قراءة الملفات النظيفة

real_estate_month = pd.read_csv(
    "C:/Users/tf/Desktop/مشروع 1/Data after cleaning/real_estate_month.csv"
)

tourism_clean = pd.read_csv(
    "C:/Users/tf/Desktop/مشروع 1/Data after cleaning/tourism_cleaned.csv"
)
```

```python

#  فحص سريع قبل الدمج

print("Real Estate shape:", real_estate_month.shape)
print("Tourism shape:", tourism_clean.shape)

print("Real Estate columns:")
print(real_estate_month.columns)

print("Tourism columns:")
print(tourism_clean.columns)
```

```python
# توحيد مفاتيح الدمج 

# year_month → YYYY-MM في الملفين
real_estate_month['year_month'] = (
    pd.to_datetime(real_estate_month['year_month'])
    .dt.to_period('M')
    .astype(str)
)

tourism_clean['year_month'] = (
    pd.to_datetime(tourism_clean['year_month'])
    .dt.to_period('M')
    .astype(str)
)

# توحيد أسماء المناطق
real_estate_month['area_name_en'] = (
    real_estate_month['area_name_en']
    .astype(str).str.strip().str.lower()
)

tourism_clean['area_name_en'] = (
    tourism_clean['area_name_en']
    .astype(str).str.strip().str.lower()
)
```

```python

#  الدمج (شهري + منطقة)

merged_df = pd.merge(
    real_estate_month,
    tourism_clean,
    on=['year_month', 'area_name_en'],
    how='left'
)

```

```python

#  فحص بعد الدمج

print("Merged shape:", merged_df.shape)
print(merged_df.head())
print(merged_df.info())

print("نسبة القيم الفارغة القادمة من السياحة (%):")
print(
    merged_df[['hotels', 'rooms', 'POIs', 'occupancy_rate']]
    .isna()
    .mean() * 100
)
```

```python

# 5) حفظ الملف النهائي

merged_df.to_csv(
    "C:/Users/tf/Desktop/مشروع 1/Data after cleaning/real_estate_tourism_merged.csv",
    index=False
)

print("✔ تم إنشاء ملف الدمج النهائي بنجاح")
```
